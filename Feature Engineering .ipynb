```python
# let's import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Let's load the dataset 
columns=['loan_amnt', 'grade', 'purpose', 'issue_d', 'last_pymnt_d']
Loan_dataset=pd.read_csv("loan.csv", usecols=columns)
```

```python
# Let's have a glance on the dataset
Loan_dataset.head(4)   
# Here issue_d and last_pamnt_d has dates in an order that needs to be converted into acceptable form.
```

```python
Loan_dataset.dtypes   # To check the data types of the variables
```

```python
# Lets' convert them into pandas date_time format
Loan_dataset["issue_dt"]=pd.to_datetime(Loan_dataset.issue_d)
Loan_dataset["last_pymnt_dt"]=pd.to_datetime(Loan_dataset.last_pymnt_d)
```

```python
Loan_dataset[["issue_dt", "issue_d", "last_pymnt_dt", "last_pymnt_d"]].head()   # comparing before and after
```

```python
#Let's plot the dates vs loan amount
fig=Loan_dataset.groupby(["issue_dt", "grade"])["loan_amnt"].sum().unstack().plot(figsize=(14,10))
fig.set_title('Disbursed amount in time')
fig.set_ylabel('Disbursed Amount (US Dollars)')
```
